```python
import cv2
import numpy as np
import argparse
from tkinter import * 
from PIL import Image, ImageTk
```

```python
img = cv2.imread('files/laby_rgb.png')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
## get only the blue color
lower_blue = np.array([110,50,50])  
upper_blue = np.array([130,255,255])

# get only the red color
lower_red = np.array([0,50,50])
upper_red = np.array([10,255,255])

# get only the green color
lower_green = np.array([50,50,50])
upper_green = np.array([70,255,255])
```

```python
root = Tk()
root.geometry("1800x1500")
root.title("Drawing lines to a canvas")
cv = Canvas(root,height="1800",width="1500",bg="white")
cv.pack()
```

```python
list_coord_contour = []
list_coord_contour_red = []
list_coord_contour_green = []


# show mask laby
mask_laby = cv2.inRange(hsv, lower_blue, upper_blue)
#show contour of img laby
edges_laby = cv2.Canny(mask_laby,100,200)

# show mask red 
mask_red = cv2.inRange(hsv, lower_red, upper_red)
#show contour of img red
edges_red = cv2.Canny(mask_red,100,200)

# show mask green
mask_green = cv2.inRange(hsv, lower_green, upper_green)
#show contour of img green
edges_green = cv2.Canny(mask_green,100,200)


#get coords of canny detection red
contours_red, hierarchy = cv2.findContours(edges_red, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours_red)):
    x,y,w,h = cv2.boundingRect(contours_red[i])
    list_coord_contour.append([x,y,w,h])
    cv.create_rectangle(x,y,x+w,y+h,fill="red")
    list_coord_contour_red.append([x,y,w,h])

#get coords of canny detection green
contours_green, hierarchy = cv2.findContours(edges_green, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours_green)):
    x,y,w,h = cv2.boundingRect(contours_green[i])
    cv.create_rectangle(x,y,x+w,y+h,fill="green")
    list_coord_contour_green.append([x,y,w,h])
    


#get coords of canny detection
contours, hierarchy = cv2.findContours(edges_laby, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    list_coord_contour.append((x,y,w,h))



for i in range(len(list_coord_contour)):
    cv.create_rectangle(list_coord_contour[i][0], list_coord_contour[i][1], list_coord_contour[i][0]+list_coord_contour[i][2], list_coord_contour[i][1]+list_coord_contour[i][3], outline = "black", fill = "white", width = 2)

for i in range(len(list_coord_contour_red)):
    cv.create_rectangle(list_coord_contour_red[i][0], list_coord_contour_red[i][1], list_coord_contour_red[i][0]+list_coord_contour_red[i][2], list_coord_contour_red[i][1]+list_coord_contour_red[i][3], outline = "black", fill = "white", width = 2)

for i in range(len(list_coord_contour_green)):
    cv.create_rectangle(list_coord_contour_green[i][0], list_coord_contour_green[i][1], list_coord_contour_green[i][0]+list_coord_contour_green[i][2], list_coord_contour_green[i][1]+list_coord_contour_green[i][3], outline = "black", fill = "white", width = 2)

oval = cv.create_oval(list_coord_contour_green[0][0], list_coord_contour_green[0][1], list_coord_contour_green[0][0]+list_coord_contour_green[0][2], list_coord_contour_green[0][1]+list_coord_contour_green[0][3], outline = "black", fill = "white", width = 2)

print(list_coord_contour_green)


def colision_between_oval_and_rect():
    for i in range(len(list_coord_contour)):
        if (cv.coords(oval)[0] >= list_coord_contour[i][0] and cv.coords(oval)[0] <= list_coord_contour[i][0]+list_coord_contour[i][2] and cv.coords(oval)[1] >= list_coord_contour[i][1] and cv.coords(oval)[1] <= list_coord_contour[i][1]+list_coord_contour[i][3]):
            
            print(cv.coords(oval)[0],list_coord_contour[i][0])
            print("collision")
            return True
    return False
80


def left(event):
    x = -10
    y = 0
    cv.move(oval, x, y)
    if (colision_between_oval_and_rect()):
        x = 50
        y = 0
        cv.move(oval, x, y)

        print(colision_between_oval_and_rect())

def right(event):
    x = 10
    y = 0
    cv.move(oval, x, y)
    if (colision_between_oval_and_rect()):
        x = -50
        y = 0
        cv.move(oval, x, y)
        print(colision_between_oval_and_rect())

def up(event):
    x = 0
    y = -10
    cv.move(oval, x, y)
    if (colision_between_oval_and_rect()):
        x = 0
        y = 50
        cv.move(oval, x, y)
        print(colision_between_oval_and_rect())

def down(event):
    x = 0
    y = 10
    cv.move(oval, x, y)
    if (colision_between_oval_and_rect()):
        x = 0
        y = -50
        cv.move(oval, x, y)
        print(colision_between_oval_and_rect())



root.bind("<Left>", left)
root.bind("<Right>", right)
root.bind("<Up>", up)
root.bind("<Down>", down)


root.mainloop() 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
